```python
with open('./output.txt', 'r') as f:
    t = f.read()
for k in ['embed_size', ' model', 'sampled_neighbours', 'lr', 'num_negative_samples']:
    t = t.replace(k, f"\'{k}\'")


t = t.replace("<class '__main__.Model'>", "\'ModelClass\'")
t = t.replace("<class '__main__.ModelSimple'>", "\'ModelSimpleClass\'")
```

```python
with open('./output_refactored.txt', 'w') as f:
    f.write(t)
```

```python
with open('./output_refactored.txt', 'r') as f:
    lines = [l.strip() for l in f.readlines()]
    results = []
    curr_res = {}
    for line in lines[1:]:
        if line.startswith('\'embed'):
            #print('{' + line + '}')
            curr_res = eval('{' + line + '}')
            
        if line.startswith('precision') or line.startswith('recall'):
            k, v = line.split(':')[0], float(line.split(':')[1])
            curr_res[k] = v 
        if line.startswith('Proxy'):
            results.append(curr_res)
        
```

```python
for r in results:
    r['sampled_neighbours'] = r['sampled_neighbours'][(('concept', 'rev_course_concept', 'course'))][0]
```
